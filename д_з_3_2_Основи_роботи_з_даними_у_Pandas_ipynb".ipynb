9. Врахуовуючи висновок в попередньому завданні, дослідимо як продаються в цих містах різні категорії товарів. Виведіть відсотковий розпоіділ кількості інвойсів по категоріям товарів `Product line` в містах `Yangon` та `Mandalay`.

```python
# Для Yangon
yangon_distribution = df[df['City'] == 'Yangon']['Product line'].value_counts(normalize=True) * 100
print("Відсотковий розподіл по категоріям у Yangon:")
print(yangon_distribution)
print()

# Для Mandalay
mandalay_distribution = df[df['City'] == 'Mandalay']['Product line'].value_counts(normalize=True) * 100
print("Відсотковий розподіл по категоріям у Mandalay:")
print(mandalay_distribution)
```
